# 5. Compute Order 12 Autocorrelation Vector 

```python
auto_correlation_vector = []
m_order = 12

for i in range(len(hpsn)):
    auto_correlation_vector.append(correlation.crossCorrelationVector(hpsn[i], m_order, window_size - m_order))
    
c = 1
title = "5. Vector de autocorrelacin de orden 12  ventana ({0}/{1})".format(c, len(auto_correlation_vector))
utils.beautyPlot(auto_correlation_vector[c - 1],title)
                 
```

```python
c = len(auto_correlation_vector)
title = "5. Vector de autocorrelacin de orden 12  ({0}/{1})".format(
    c, len(auto_correlation_vector))
utils.beautyPlot(auto_correlation_vector[c - 1], title)
```

```python
print(len(auto_correlation_vector))
f, axis = plt.subplots(3, 3)
c = 1
title = "5. Seal 'S' - Vectores de autocorrelacin para cada una de las {0} ventanas ".format(
    len(auto_correlation_vector))
f.suptitle(title, fontsize=16)
axis[0, 0].plot(auto_correlation_vector[0][1:])

axis[0, 0].set_title('[1/71]')
axis[0, 1].plot(auto_correlation_vector[1][1:])
axis[0, 1].set_title('[2/71]')
axis[0, 2].plot(auto_correlation_vector[2][1:])
axis[0, 2].set_title('[3/71]')
axis[1, 0].plot(auto_correlation_vector[3][1:])
axis[1, 0].set_title('[4/71]')
axis[1, 1].plot(auto_correlation_vector[4][1:])
axis[1, 1].set_title('[5/71]')
axis[1, 2].plot(auto_correlation_vector[5][1:])
axis[1, 2].set_title('[6/71]')
axis[2, 0].plot(auto_correlation_vector[6][1:])
axis[2, 0].set_title('[7/71]')
axis[2, 1].plot(auto_correlation_vector[7][1:])
axis[2, 1].set_title('[8/71]')
axis[2, 2].plot(auto_correlation_vector[69][1:])
axis[2, 2].set_title('[70/71]')
plt.show
```

# 6. Calculando los coeficientes de Wiener

```python
auto_correlation_matrixes = [correlation.createAutoCorrelationMatrix(auto_correlation_vector[i], 12) for i in range(len(auto_correlation_vector))]
```

```python
# utils.pprintm(auto_correlation_matrixes[5])
utils.beautyPlot(auto_correlation_matrixes[5], '6. Coeficientes de Wiener - Auto Correlation Matrix [5/9]')
```

```python
auto_correlation_matrixes_inverted = [np.linalg.inv(
    auto_correlation_matrixes[i]) for i in range(len(hpsn))]

auto_correlation_vector_matrix = [np.array(auto_correlation_vector[i]).reshape(
    len(auto_correlation_vector[i]), 1) for i in range(len(hpsn))]

wiener_coefficients_per_window = [
    auto_correlation_matrixes_inverted[i]@auto_correlation_vector_matrix[i] for i in range(len(hpsn))]
```

```python
print(len(wiener_coefficients_per_window))


# utils.beautyPlot(wiener_coefficients_per_window[5], '6. Coeficientes de Wiener [5/71]')

f, axis = plt.subplots(3, 3)
c = 1
title = "6. Seal 'S' - Coeficientes de Wiener para las {0} ventanas ".format(
    len(auto_correlation_vector))
f.suptitle(title, fontsize=16)
axis[0, 0].plot(wiener_coefficients_per_window[0][1:])

axis[0, 0].set_title('[1/71]')
axis[0, 1].plot(wiener_coefficients_per_window[1][1:])
axis[0, 1].set_title('[2/71]')
axis[0, 2].plot(wiener_coefficients_per_window[2][1:])
axis[0, 2].set_title('[3/71]')
axis[1, 0].plot(wiener_coefficients_per_window[3][1:])
axis[1, 0].set_title('[4/71]')
axis[1, 1].plot(wiener_coefficients_per_window[4][1:])
axis[1, 1].set_title('[5/71]')
axis[1, 2].plot(wiener_coefficients_per_window[5][1:])
axis[1, 2].set_title('[6/71]')
axis[2, 0].plot(wiener_coefficients_per_window[6][1:])
axis[2, 0].set_title('[7/71]')
axis[2, 1].plot(wiener_coefficients_per_window[7][1:])
axis[2, 1].set_title('[8/71]')
axis[2, 2].plot(wiener_coefficients_per_window[69][1:])
axis[2, 2].set_title('[70/71]')
plt.show
```
